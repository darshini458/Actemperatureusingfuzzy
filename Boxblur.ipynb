```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread('/content/cat image.webp')
```

```python
plt.subplot(1,2,1)
plt.imshow(image)
plt.show()
```

```python
filtered_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,2)
plt.imshow(filtered_image)
plt.show()
```

```python
kernal=np.ones((3,3),np.float32)/9
new_image=cv2.filter2D(filtered_image,-1,kernal)
plt.subplot(1,2,2)
plt.imshow(new_image)
plt.show()
```

```python
kernal2=np.ones((3,3),np.float32)/9
new_image2=cv2.filter2D(filtered_image,-1,kernal2)
fig,ax=plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(new_image2)
ax[1].imshow(filtered_image)
plt.show()
```

```python
brightness=10
contrast=2
new_image3=cv2.addWeighted(filtered_image,contrast,np.zeros(filtered_image.shape,filtered_image.dtype),0,brightness)
plt.subplot(1,2,2)
plt.imshow(new_image3)
plt.show()
```

```python
kernal3=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
new_image4=cv2.filter2D(filtered_image,-1,kernal3)
plt.subplot(1,2,2)
plt.imshow(new_image4)
plt.show()
```

```python
kernal4=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
new_image5=cv2.filter2D(filtered_image,-1,kernal4)
plt.subplot(1,2,2)
plt.imshow(new_image5)
plt.show()
```

```python
kernal5=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
new_image6=cv2.filter2D(filtered_image,-1,kernal5)
plt.subplot(1,2,2)
plt.imshow(new_image6)
plt.show()
```
